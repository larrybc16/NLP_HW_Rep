Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin.

```python
# Initial imports
import os
import pandas as pd
from dotenv import load_dotenv
from nltk.sentiment.vader import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

%matplotlib inline
```

```python
# Fetch the Ethereum news articles
# YOUR CODE HERE!
eth_art = newsapi.get_everything(q="ethereum")
```

```python
# Create the Bitcoin sentiment scores DataFrame
# YOUR CODE HERE!


lst_bit = []
for articles in bit_art["articles"]:
 
    if  articles["content"]: 
        txt = articles["content"]
        results = analyzer.polarity_scores(txt)
        res = results["compound"]
        pos = results["pos"]
        nuet = results["neu"]
        neg = results["neg"]
        lst_bit.append({"text":txt, "com":res, "pos":pos, "nuet":nuet, "neg":neg})

bsen_df = pd.DataFrame(lst_bit)
bsen_df
```

```python
# Create the ethereum sentiment scores DataFrame
# YOUR CODE HERE!

lst_eth = []
for articles in eth_art["articles"]:
 
    if  articles["content"]: 
        txt = articles["content"]
        results = analyzer.polarity_scores(txt)
        res = results["compound"]
        pos = results["pos"]
        nuet = results["neu"]
        neg = results["neg"]
        lst_eth.append({"text":txt, "com":res, "pos":pos, "nuet":nuet, "neg":neg})

esen_df = pd.DataFrame(lst_eth)
esen_df
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    words = word_tokenize(text)
    sw = set(stopwords.words('english'))
    #regex = re.compile("[^a-zA-Z ]")
    sw_addons = {}
    #run once and then I will notice stop words will come up dont have to put in anything
    
    
    #re_clean = regex.sub('', words)
    #words = word_tokenize(re_clean)
    output = [word.lower() for word in words if word.lower() not in sw.union(sw_addons)]
    # Create a list of the words
    

        
        
   
    

    # Convert the words to lowercase (Correct)
    words = [word.lower() for word in words]
    
    # Remove the punctuation
    #https://stackoverflow.com/questions/15547409/how-to-get-rid-of-punctuation-using-nltk-tokenizer/56847275#56847275
    
    words = list(filter(lambda word: word not in punctuation, words))
    
    # Remove the stop words
   
    words = [word for word in words if word not in sw]
    
    # Lemmatize Words into root words
    lemmatizer = WordNetLemmatizer()
    #words = [lemmatizer.lemmatize(word) for word in words]
    #tokens = [word.lower() for word in lem if word.lower() not in sw]
    
    tokens = [lemmatizer.lemmatize(word) for word in words]
    return tokens
```

```python
# Create a new tokens column for bitcoin
# YOUR CODE HERE!
bsen_df['tokens'] = bsen_df['text'].apply(tokenizer)
bsen_df
```

```python
# Create a new tokens column for ethereum
# YOUR CODE HERE!
esen_df['tokens'] = esen_df['text'].apply(tokenizer)
esen_df
```
